# Meaningful visualizations using the dataset



```python
grouped_data = data.groupby('Classification').mean()
```

```python
grouped_data.plot(kind='bar')
plt.title('Mean Values by Classification')
plt.show()
```

```python
# Plot the data using scatter plot
sns.scatterplot(x='SL', y='SW', hue='Classification', data=data)
```

```python
sns.scatterplot(x='PL', y='PW',hue='Classification',  data=data)
```

```python
# Create a histogram of sepal length
plt.hist(data['SL'], bins=20,rwidth=0.96)
plt.xlabel('Sepal length')
plt.ylabel('Frequency')
plt.title('Histogram of Sepal Length')
plt.show()
```
